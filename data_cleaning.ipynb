### Engine Volume ###

```python
step1 = ['year', 'car', 'model']
step2 = ['year',  'car', 'Body Style']
step3 = ['year', 'car']
```

```python
# Group by specified columns and transform 'Engine Volume' column by filling missing values with median of each group
data['Engine Volume'] = data.groupby(step1)['Engine Volume'].transform(lambda x: x.fillna(x.median()) if x.notnull().any() else x)
data['Engine Volume'].isnull().sum()
```

```python
data['Engine Volume'] = data.groupby(step2)['Engine Volume'].transform(lambda x: x.fillna(x.median()) if x.notnull().any() else x)
data['Engine Volume'].isnull().sum()
```

```python
data['Engine Volume'] = data.groupby(step3)['Engine Volume'].transform(lambda x: x.fillna(x.median()) if x.notnull().any() else x)
data['Engine Volume'].isnull().sum()
```

```python
data['Engine Volume'] = data['Engine Volume'].fillna(data['Engine Volume'].median())
data['Engine Volume'].isnull().sum()
```

### Horsepower ###

Converting the Horsepower column to integer values and filling the NaN values according to existing data.

```python
data['Horsepower'] = data['Horsepower'].astype(str).str.extract(r'(\d+)').astype(float)
```

```python
data['Horsepower'] = data.groupby(step1)['Horsepower'].transform(lambda x: x.fillna(x.median()) if x.notnull().any() else x)
data['Horsepower'].isnull().sum()
```

Since there are still NA values, let's find matches by just year and make.

```python
data['Horsepower'] = data.groupby(step2)['Horsepower'].transform(lambda x: x.fillna(x.median()) if x.notnull().any() else x)
data['Horsepower'].isnull().sum()
```

```python
data['Horsepower'] = data.groupby(step3)['Horsepower'].transform(lambda x: x.fillna(x.median()) if x.notnull().any() else x)
data['Horsepower'].isnull().sum()
```

Now, let's fill in with the median value.

```python
data['Horsepower'] = data['Horsepower'].fillna(data['Horsepower'].median()).astype(int)
data['Horsepower'].isnull().sum()
```

Done! No more missing values for horsepower.

### Expressing the price as dolars and mileage as km ###

```python
data['price'] = data['price'].str.strip()
test = data['price'].str[-1]
test.unique()
```

```python
def to_dol_converter(x):
    if '₽' in x:
        return int(''.join(filter(str.isdigit, x))) * 0.011
    elif '€' in x:
        return int(''.join(filter(str.isdigit, x))) * 0.92
    elif '֏' in x:
        return int(''.join(filter(str.isdigit, x))) * 0.0025
    elif '$' in x:
        return int(''.join(filter(str.isdigit, x)))
    

data['price'] = data['price'].apply(to_dol_converter)
```

Convert miles to km.

```python
data['Mileage'] = data['Mileage'].apply(lambda x: int(x.split(' ')[0]) * 1.6 if 'miles' in x else int(x.split(' ')[0]))
data
```
